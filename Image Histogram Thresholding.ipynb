```python
#Importing dependecies
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('sample.webp')
plt.imshow(img)
print(img.shape)
```

```python
#Converting image to greyscale
gray_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2.imshow("Greyscale Image",gray_image)
cv2.waitKey(0)
cv2.destoryAllWindows()
```

```python
#Finding the histogram of the greyscale image
plt.hist(gray_image.ravel(),256,[0,256])
plt.show()
```

```python
#Getting threshold value from user
th = input("Please enter a threshold value for binary conversion: ")
th_no = int(th)
```

```python
(th, bw) = cv2.threshold(gray_image, th_no, 255, cv2.THRESH_BINARY)
```

```python
#showing image after binary conversion with threshold equals to th
cv2.imshow("Binary Image",bw)
cv2.waitKey(0)
cv2.distroyAllWindows()
```
